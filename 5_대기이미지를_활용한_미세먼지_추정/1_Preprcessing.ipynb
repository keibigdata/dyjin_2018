```python
from sklearn import datasets 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import os  
import math
%matplotlib inline

# 메타데이터 로드

meta = pd.read_csv('./data/all.csv',encoding='CP949')
PM10 = meta['PM10']
PM25 = meta['PM25']

# PM10 기초 통계
PM10 = PM10[PM10.notna()]
sns.distplot(PM10)

print(PM10.describe())

# PM25 기초 통계
PM25 = PM25[PM25.notna()]
sns.distplot(PM25)

print(PM25.describe())

# 이미지 매핑, 전처리, 트레이닝_데이터
result_mat = list()

file_list = os.listdir("./train_data/")

meta_date = meta['측정일시']

i = 0
for f in file_list:
    #print(i)
    temp = f.split(".")[0]
    temp = temp.split(" ")
    date = "".join(temp[0].split("-")) + temp[1]
    
    pm10 = list(meta.loc[meta_date == int(date),'PM10'].values)[0]
    pm25 = list(meta.loc[meta_date == int(date),'PM25'].values)[0]
    
    if math.isnan(pm10):
        continue
    if math.isnan(pm25):
        continue
        
    f = "./train_data/" + f
    result = [f,date,pm10,pm25]
    result_mat.append(result)
    i = i+1
    
result_mat = pd.DataFrame(result_mat)
result_mat.columns = ["file","date","pm10","pm25"]
result_mat.to_csv("./result/train.csv",index=None)  
```
